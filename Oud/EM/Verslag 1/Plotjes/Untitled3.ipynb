```r
library("ggplot2")
```

```r
plotname <- "part_2_withcoax"
txtfile <- paste(plotname, ".txt", sep="")
```

```r
filename <- paste("~/Temp/", txtfile, sep="")
```

```r
zl <- read.csv(file=filename, sep='\t')
zl$X <- NULL
names(zl)[names(zl)=="Ch1.mV."] <- "mV"
names(zl)[names(zl)=="X.nS."] <- "nS"
```

```r
ggplot(zl, aes(nS, mV)) + geom_line()
```

```r
zlfig <- ggplot(zl, aes(nS, mV)) + geom_line() +  scale_y_continuous(breaks=seq(-140,140,20), limits=c(-145,145)) 
```

```r
zlfig
```

```r
ggsave(paste(plotname, ".pdf", sep=""))
```
